# Scraping the topics on GitHub
-we will scrape https://github.com/topics
-we will get a list of topics. For each topic, we will get topic tittle, topic page Url and topic description
-For each topic we will create a CSV file in the following format:'title','description','url'.

# Use Beautifulsoup to parse and extract information

```python
!pip install beautifulsoup4 
```

```python
from bs4 import BeautifulSoup
```

```python
doc=BeautifulSoup(page_contents, 'html.parser')
```

```python
topic_title_p_tags=doc.find_all('p',{'class':"f3 lh-condensed mb-0 mt-1 Link--primary"})
```

```python
topic_title_p_tags
```

```python
topic_description_p_tags=doc.find_all('p',{'class':"f5 color-text-secondary mb-0 mt-1"})
topic_description_p_tags
```

```python
topic_link_tags=doc.find_all('a',{'class':"d-flex no-underline" })
len(topic_link_tags)
```

```python
topic_url="https://github.com"+topic_link_tags[0]['href']
print(topic_url)
```

```python
topic_titles=[]
topic_descriptions=[]
topic_urls=[]
base_url='https://github.com'

for tag in topic_title_p_tags:
    topic_titles.append(tag.text)

for tag in topic_description_p_tags:
    topic_descriptions.append(tag.text.strip())
    
for tag in topic_link_tags:
    topic_urls.append(base_url+tag['href'])    
    
```

```python
print(topic_titles)
```

```python
print(topic_descriptions)
```

```python
print(topic_urls)
```

```python
import pandas as pd
```

```python
import datetime
```

```python
topic_df=pd.DataFrame(list(zip(topic_titles,topic_descriptions,topic_urls)),columns=['title','description','url'])
```

```python
topic_df['timestamp'] = ct
```
